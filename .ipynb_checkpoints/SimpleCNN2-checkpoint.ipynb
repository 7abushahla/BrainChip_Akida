```python
def create_simplecnn2_functional(num_classes=12, input_shape=(480, 440, 1)):
    """
    SimpleCNN2 using Functional API (required for Akida quantization)
    """
    # Input layer
    inputs = keras.Input(shape=input_shape, name='input')
    
    # Conv block 1
    x = keras.layers.Conv2D(16, kernel_size=5, name='conv1')(inputs)
    x = keras.layers.ReLU(max_value=6.0, name='relu1')(x)  # Bounded ReLU for Akida v1
    x = keras.layers.MaxPool2D(pool_size=2, strides=2, name='maxpool1')(x)
    
    # Conv block 2
    x = keras.layers.Conv2D(16, kernel_size=5, name='conv2')(x)
    x = keras.layers.ReLU(max_value=6.0, name='relu2')(x)
    x = keras.layers.MaxPool2D(pool_size=2, strides=2, name='maxpool2')(x)
    
    # Conv block 3
    x = keras.layers.Conv2D(16, kernel_size=5, name='conv3')(x)
    x = keras.layers.ReLU(max_value=6.0, name='relu3')(x)
    x = keras.layers.MaxPool2D(pool_size=2, strides=2, name='maxpool3')(x)
    
    # Conv block 4
    x = keras.layers.Conv2D(32, kernel_size=5, name='conv4')(x)
    x = keras.layers.ReLU(max_value=6.0, name='relu4')(x)
    x = keras.layers.MaxPool2D(pool_size=2, strides=2, name='maxpool4')(x)
    
    # Global pooling
    # x = keras.layers.GlobalAveragePooling2D(name='avgpool')(x)
    x = keras.layers.Flatten(name='flatten')(x)  # Use Flatten instead of GAP
    
    # Fully connected layers
    x = keras.layers.Dense(128, name='fc1')(x)
    x = keras.layers.ReLU(max_value=6.0, name='relu_fc1')(x)
    
    x = keras.layers.Dense(64, name='fc2')(x)
    x = keras.layers.ReLU(max_value=6.0, name='relu_fc2')(x)
    
    outputs = keras.layers.Dense(num_classes, name='fc3')(x)
    
    # Create model
    model = keras.Model(inputs=inputs, outputs=outputs, name='SimpleCNN2')
    
    return model
```

```python
def create_simplecnn2_model(num_classes=12, input_shape=(440, 440, 1), kernel_size=5):
    """
    SimpleCNN2 - Akida v1 compatible version
    Key changes for Akida v1:
    - Separate activation layers (not fused with Conv/Dense)
    - BatchNorm after conv layers for better quantization
    - Flatten instead of GlobalAveragePooling2D (Akida v1 compatibility)
    - InputConv: no padding (valid) - Akida allows 'same' or 'valid'
    - Intermediate Conv: padding='same' (REQUIRED for CNP hardware acceleration!)
    - Bounded ReLU with max_value=6.0 (Akida v1 requirement)
    """
    model = keras.Sequential([
        # Input layer
        keras.layers.Input(shape=input_shape),
        
        # Conv block 1 (InputConvolutional): Can use 'valid' or 'same'
        keras.layers.Conv2D(16, kernel_size=kernel_size, padding='valid'),
        keras.layers.BatchNormalization(),
        keras.layers.ReLU(max_value=6.0),
        keras.layers.MaxPool2D(pool_size=2, strides=2, padding='valid'),  # Must match Conv
        
        # Conv block 2 (Convolutional): MUST use 'same' for hardware acceleration
        keras.layers.Conv2D(16, kernel_size=kernel_size, padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.ReLU(max_value=6.0),
        keras.layers.MaxPool2D(pool_size=2, strides=2, padding='same'),  # Must match Conv
        
        # Conv block 3 (Convolutional): MUST use 'same' for hardware acceleration
        keras.layers.Conv2D(16, kernel_size=kernel_size, padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.ReLU(max_value=6.0),
        keras.layers.MaxPool2D(pool_size=2, strides=2, padding='same'),  # Must match Conv
        
        # Conv block 4 (Convolutional): NO MaxPool - last conv cannot have pooling before Dense
        keras.layers.Conv2D(32, kernel_size=kernel_size, padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.ReLU(max_value=6.0),
        # NO MaxPool here - Akida requires conv with pooling to be followed by another conv
        
        # Flatten
        keras.layers.Flatten(),
        
        # Fully connected layers with separate bounded activations
        keras.layers.Dense(128),
        keras.layers.ReLU(max_value=6.0),
        
        keras.layers.Dense(64),
        keras.layers.ReLU(max_value=6.0),
        
        keras.layers.Dense(num_classes)
    ], name='SimpleCNN2')
    
    return model
```

```python
def load_data(base_path='./minimum viable', num_files=26):
    """Load and preprocess the .npy event data files"""
    paths = [f'{base_path}/misalignment_x_combined_events_{i:04d}.npy' for i in range(num_files)]
    
    samples_float = []
    samples_uint8 = []
    print(f"Loading {num_files} samples...")
    
    for i, path in enumerate(paths):
        sample_npy = np.load(path)[:, 130:570]
        max_val = np.max(sample_npy)
        if max_val == 0:
            img_normalized = sample_npy.astype(np.float32)
        else:
            img_normalized = sample_npy.astype(np.float32) / max_val
        
        # Add channel dimension (TensorFlow uses channels-last format)
        img_tensor_float = np.expand_dims(img_normalized, axis=-1)
        samples_float.append(img_tensor_float)
        
        # Also create uint8 version for Akida (scale to 0-255)
        img_uint8 = (img_normalized * 255).astype(np.uint8)
        img_tensor_uint8 = np.expand_dims(img_uint8, axis=-1)
        samples_uint8.append(img_tensor_uint8)
    
    # Stack into single arrays
    samples_float_array = np.stack(samples_float, axis=0)
    samples_uint8_array = np.stack(samples_uint8, axis=0)
    print(f"Loaded data shape: {samples_float_array.shape}")
    print(f"Float32 data range: [{samples_float_array.min():.3f}, {samples_float_array.max():.3f}]")
    print(f"Uint8 data range: [{samples_uint8_array.min()}, {samples_uint8_array.max()}]")
    
    return samples_float_array, samples_uint8_array
```

```python
def benchmark_keras_model(model, samples):
    """Benchmark the Keras model on CPU/GPU"""
    print("\n" + "="*70)
    print("KERAS MODEL BENCHMARKING (CPU/GPU)")
    print("="*70)
    
    proc_time = []
    for i in range(len(samples)):
        sample = samples[i:i+1]
        start_time = time.time()
        _ = model(sample)
        end_time = time.time()
        
        # Skip first iteration (warmup)
        if i == 0:
            continue
        proc_time.append(1000 * (end_time - start_time))
    
    mean_time = np.mean(proc_time)
    std_time = np.std(proc_time)
    
    print(f"Number of samples: {len(samples)-1}")
    print(f"Average inference time: {mean_time:.3f} ± {std_time:.3f} ms")
    print(f"Throughput: {1000/mean_time:.1f} fps")
    print("="*70)
    
    return mean_time, std_time


def benchmark_akida_hardware(model_akida, device, samples):
    """Comprehensive benchmarking of Akida hardware performance"""
    
    print("\n" + "="*70)
    print("AKIDA HARDWARE BENCHMARKING")
    print("="*70)
    
    # Enable power measurement
    device.soc.power_measurement_enabled = True
    
    # Store results for summary
    results = {}
    
    # 1. Batch inference with forward() - best for throughput and power measurement
    print("\n1. BATCH INFERENCE (forward method):")
    print("-"*70)
    
    start = time.perf_counter()
    _ = model_akida.forward(samples)
    end = time.perf_counter()
    
    total_time = (end - start) * 1000  # ms
    avg_time = total_time / len(samples)
    
    # Extract statistics
    stats_str = str(model_akida.statistics)
    
    fps_match = re.search(r'Average framerate = ([\d.]+) fps', stats_str)
    hw_fps = float(fps_match.group(1)) if fps_match else None
    
    power_match = re.search(r'power range \(mW\):\s+Avg\s+([\d.]+)', stats_str)
    avg_power = float(power_match.group(1)) if power_match else None
    
    energy_match = re.search(r'energy consumed \(mJ/frame\):\s+([\d.]+)', stats_str)
    energy = float(energy_match.group(1)) if energy_match else None
    
    floor_power = device.soc.power_meter.floor
    
    # Store batch results
    results['batch_time_ms'] = avg_time
    results['batch_fps'] = 1000/avg_time
    results['avg_power_mw'] = avg_power
    results['energy_mj'] = energy
    results['floor_power_mw'] = floor_power
    
    print(f"Total samples: {len(samples)}")
    print(f"Total time: {total_time:.2f} ms")
    print(f"Avg time per sample: {avg_time:.3f} ms")
    print(f"Measured framerate: {1000/avg_time:.2f} fps")
    if hw_fps:
        print(f"Hardware framerate: {hw_fps:.2f} fps")
    print(f"\nFloor power: {floor_power:.2f} mW")
    if avg_power:
        print(f"Average power: {avg_power:.2f} mW")
    if energy:
        print(f"Energy per inference: {energy:.2f} mJ")
        print(f"Energy efficiency: {1000/energy:.0f} inferences/joule")
    if avg_power:
        print(f"Power efficiency: {(1000/avg_time)/(avg_power/1000):.0f} fps/watt")
    
    # 2. Single-image inference timing
    print("\n2. SINGLE-IMAGE INFERENCE (predict method):")
    print("-"*70)
    
    inference_times = []
    num_runs = min(100, len(samples))
    
    # Warmup
    _ = model_akida.predict(samples[0:1])
    
    # Measure
    for i in range(num_runs):
        sample = samples[i:i+1]
        start = time.perf_counter()
        _ = model_akida.predict(sample)
        end = time.perf_counter()
        inference_times.append((end - start) * 1000)
    
    inference_times = np.array(inference_times)
    
    # Store single-image results
    results['single_time_ms'] = np.mean(inference_times)
    results['single_std_ms'] = np.std(inference_times)
    results['single_min_ms'] = np.min(inference_times)
    results['single_max_ms'] = np.max(inference_times)
    
    print(f"Runs: {num_runs}")
    print(f"Mean ± Std: {np.mean(inference_times):.3f} ± {np.std(inference_times):.3f} ms")
    print(f"Median: {np.median(inference_times):.3f} ms")
    print(f"Min: {np.min(inference_times):.3f} ms")
    print(f"Max: {np.max(inference_times):.3f} ms")
    print(f"Throughput: {1000/np.mean(inference_times):.1f} fps")
    
    # 3. Clock mode comparison
    print("\n3. CLOCK MODE COMPARISON:")
    print("-"*70)
    
    from akida.core.soc import ClockMode
    
    clock_modes = [
        (ClockMode.Performance, "Performance"),
        (ClockMode.Economy, "Economy"),
        (ClockMode.LowPower, "LowPower")
    ]
    
    clock_results = []
    test_batch = samples[:min(100, len(samples))]
    
    for mode, mode_name in clock_modes:
        device.soc.clock_mode = mode
        
        # Warmup
        _ = model_akida.forward(test_batch[:10])
        
        # Measure
        start = time.perf_counter()
        _ = model_akida.forward(test_batch)
        end = time.perf_counter()
        
        total_time_ms = (end - start) * 1000
        avg_time_ms = total_time_ms / len(test_batch)
        
        stats_str = str(model_akida.statistics)
        fps_match = re.search(r'Average framerate = ([\d.]+) fps', stats_str)
        hw_fps = float(fps_match.group(1)) if fps_match else None
        
        power_match = re.search(r'power range \(mW\):\s+Avg\s+([\d.]+)', stats_str)
        avg_power = float(power_match.group(1)) if power_match else None
        
        energy_match = re.search(r'energy consumed \(mJ/frame\):\s+([\d.]+)', stats_str)
        energy_mj = float(energy_match.group(1)) if energy_match else None
        
        clock_results.append({
            'mode': mode_name,
            'time_ms': avg_time_ms,
            'fps': 1000/avg_time_ms,
            'hw_fps': hw_fps,
            'power_mw': avg_power,
            'energy_mj': energy_mj
        })
    
    print(f"{'Mode':<15} {'Time(ms)':<12} {'FPS':<12} {'Power(mW)':<12} {'Energy(mJ)':<12} {'Efficiency(fps/W)'}")
    print("-"*70)
    for r in clock_results:
        efficiency = (r['fps'] / (r['power_mw'] / 1000)) if r['power_mw'] else 0
        print(f"{r['mode']:<15} {r['time_ms']:<12.3f} {r['fps']:<12.0f} "
              f"{r['power_mw'] if r['power_mw'] else 'N/A':<12} "
              f"{r['energy_mj'] if r['energy_mj'] else 'N/A':<12} "
              f"{efficiency:<12.0f}")
    
    # Set to Economy mode (best balance)
    device.soc.clock_mode = ClockMode.Economy
    print(f"\n✅ Clock mode set to Economy (best efficiency)")
    
    print("="*70)
    
    return results


def main_large_input():
    """
    Original pipeline with large input (480x440) - Conv layers run in SOFTWARE
    This exceeds Akida v1's 256-pixel width constraint
    """
    print("="*70)
    print("AKIDA DEPLOYMENT - LARGE INPUT (480x440) - SOFTWARE MODE")
    print("="*70)
    
    # Load data - get both float32 and uint8 versions
    samples_float, samples_uint8 = load_data()
    
    # Auto-detect input shape from data
    input_shape = samples_float.shape[1:]  # (height, width, channels)
    print(f"\nDetected input shape: {input_shape}")
    
    # Step 1: Create Akida-compatible Keras model
    print("\nStep 1: Creating Akida-compatible Keras model...")
    with set_akida_version(AkidaVersion.v1):
        model = create_simplecnn2_model(num_classes=12, input_shape=input_shape)
        
        print("\nModel Summary:")
        # model.summary()
    
    # Step 2: Benchmark Keras model (uses float32)
    print("\nStep 2: Benchmarking Keras model (baseline)...")
    keras_time, keras_std = benchmark_keras_model(model, samples_float)
    
    # Step 3: Quantize model (uses float32)
    print("\nStep 3: Quantizing model for Akida...")
    with set_akida_version(AkidaVersion.v1):
        # Start with 4-bit quantization for efficiency
        qparams = QuantizationParams(input_weight_bits=8, weight_bits=4, activation_bits=4)
        print("Quantizing with 4-bit weights and activations...")
        
        model_quantized = quantize(
            model, 
            qparams=qparams,
            samples=samples_float[:10],  # Use subset for quantization calibration
            num_samples=10,
            batch_size=2
        )
        
        print("\nQuantized Model Summary:")
        # model_quantized.summary()
    
    # Step 4: Convert to Akida format
    print("\nStep 4: Converting to Akida format...")
    with set_akida_version(AkidaVersion.v1):
        model_akida = convert(model_quantized)
        
        print("\nAkida Model Summary:")
        model_akida.summary()
    
    # Step 5: Map to Akida hardware
    print("\nStep 5: Detecting and mapping to Akida hardware...")
    devices = akida.devices()
    
    if len(devices) == 0:
        print("❌ No Akida device found!")
        print("   Model created but cannot be deployed to hardware.")
        print("   Running in software simulation mode only.")
        return
    
    device = devices[0]
    print(f"✅ Found Akida device: {device.desc}")
    print(f"   Device version: {device.version}")
    print(f"   Model IP version: {model_akida.ip_version}")
    
    # Map to hardware
    model_akida.map(device)
    print("\n✅ Model mapped to hardware!")
    
    print("\nHardware Allocation:")
    model_akida.summary()
    
    # Step 6: Benchmark on Akida hardware (uses uint8)
    print("\nStep 6: Benchmarking on Akida hardware...")
    akida_results = benchmark_akida_hardware(model_akida, device, samples_uint8)
    
    # Step 7: Save the Akida model
    print("\nStep 7: Saving Akida model...")
    output_file = "simplecnn2_akida_v1.fbz"
    model_akida.save(output_file)
    file_size = Path(output_file).stat().st_size / 1024
    print(f"✅ Model saved to: {output_file}")
    print(f"   File size: {file_size:.2f} KB")
    
    # Summary comparison with actual measured values
    print("\n" + "="*70)
    print("PERFORMANCE SUMMARY")
    print("="*70)
    print(f"\n📊 KERAS BASELINE (CPU/GPU):")
    print(f"   Inference time: {keras_time:.3f} ± {keras_std:.3f} ms")
    print(f"   Throughput:     {1000/keras_time:.1f} fps")
    
    print(f"\n🔌 AKIDA HARDWARE:")
    print(f"   Batch inference:  {akida_results['batch_time_ms']:.3f} ms per image")
    print(f"   Single inference: {akida_results['single_time_ms']:.3f} ± {akida_results['single_std_ms']:.3f} ms")
    print(f"   Throughput:       {akida_results['batch_fps']:.1f} fps (batch mode)")
    
    print(f"\n⚡ POWER & ENERGY:")
    print(f"   Floor power:      {akida_results['floor_power_mw']:.2f} mW")
    print(f"   Average power:    {akida_results['avg_power_mw']:.2f} mW")
    print(f"   Energy/inference: {akida_results['energy_mj']:.2f} mJ")
    print(f"   Energy efficiency: {1000/akida_results['energy_mj']:.0f} inferences/joule")
    
    print(f"\n⚠️  NOTE: Conv layers running in SOFTWARE (not accelerated)")
    print(f"   Only Dense layers are hardware-accelerated")
    print(f"   Large input size ({input_shape}) exceeds Akida v1 width limit (≤256)")
    print("="*70)
    
    print("\n✅ Deployment complete!")
    
    return {
        'keras_time_ms': keras_time,
        'keras_std_ms': keras_std,
        'akida_results': akida_results,
        'input_shape': input_shape
    }


def load_data_downsampled(base_path='./minimum viable', num_files=26, target_width=256):
    """
    Load and preprocess data with downsampling to fit Akida v1 constraints
    Target: width <= 256 pixels for hardware acceleration
    """
    import cv2
    paths = [f'{base_path}/misalignment_x_combined_events_{i:04d}.npy' for i in range(num_files)]
    
    samples_float = []
    samples_uint8 = []
    print(f"Loading {num_files} samples with downsampling to width={target_width}...")
    
    for i, path in enumerate(paths):
        sample_npy = np.load(path)[:, 130:570]  # Original: (480, 440)
        
        # Calculate target height to maintain aspect ratio
        original_height, original_width = sample_npy.shape
        aspect_ratio = original_height / original_width
        target_height = int(target_width * aspect_ratio)
        
        # Normalize first
        max_val = np.max(sample_npy)
        if max_val == 0:
            img_normalized = sample_npy.astype(np.float32)
        else:
            img_normalized = sample_npy.astype(np.float32) / max_val
        
        # Downsample using OpenCV
        img_resized = cv2.resize(img_normalized, (target_width, target_height), 
                                 interpolation=cv2.INTER_AREA)
        
        # Float32 version for Keras
        img_tensor_float = np.expand_dims(img_resized, axis=-1)
        samples_float.append(img_tensor_float)
        
        # Uint8 version for Akida (scale to 0-255)
        img_uint8 = (img_resized * 255).astype(np.uint8)
        img_tensor_uint8 = np.expand_dims(img_uint8, axis=-1)
        samples_uint8.append(img_tensor_uint8)
    
    # Stack into single arrays
    samples_float_array = np.stack(samples_float, axis=0)
    samples_uint8_array = np.stack(samples_uint8, axis=0)
    print(f"Downsampled data shape: {samples_float_array.shape}")
    print(f"Float32 data range: [{samples_float_array.min():.3f}, {samples_float_array.max():.3f}]")
    print(f"Uint8 data range: [{samples_uint8_array.min()}, {samples_uint8_array.max()}]")
    
    return samples_float_array, samples_uint8_array


def main_small_input():
    """
    Optimized pipeline with downsampled input (<=256 width)
    Uses 5x5 kernels (original architecture)
    """
    print("="*70)
    print("AKIDA DEPLOYMENT - SMALL INPUT (≤256 width) - 5x5 KERNELS")
    print("="*70)
    
    # Load downsampled data
    samples_float, samples_uint8 = load_data_downsampled(target_width=220)
    
    # Auto-detect input shape from data
    input_shape = samples_float.shape[1:]  # Should be around (240, 220, 1)
    print(f"\nDetected input shape: {input_shape}")
    print(f"✅ Width {input_shape[1]} <= 256 (Akida v1 constraint satisfied)")
    
    # Step 1: Create Akida-compatible Keras model with 5x5 kernels
    print("\nStep 1: Creating Akida-compatible Keras model (5x5 kernels)...")
    with set_akida_version(AkidaVersion.v1):
        model = create_simplecnn2_model(num_classes=12, input_shape=input_shape, kernel_size=5)
        
        print("\nModel Summary:")
        # model.summary()
    
    # Step 2: Benchmark Keras model (uses float32)
    print("\nStep 2: Benchmarking Keras model (baseline)...")
    keras_time, keras_std = benchmark_keras_model(model, samples_float)
    
    # Step 3: Quantize model (uses float32)
    print("\nStep 3: Quantizing model for Akida...")
    with set_akida_version(AkidaVersion.v1):
        # Start with 4-bit quantization for efficiency
        qparams = QuantizationParams(input_weight_bits=8, weight_bits=4, activation_bits=4)
        print("Quantizing with 4-bit weights and activations...")
        
        model_quantized = quantize(
            model, 
            qparams=qparams,
            samples=samples_float[:10],  # Use subset for quantization calibration
            num_samples=10,
            batch_size=2
        )
        
        print("\nQuantized Model Summary:")
        # model_quantized.summary()
    
    # Step 4: Convert to Akida format
    print("\nStep 4: Converting to Akida format...")
    with set_akida_version(AkidaVersion.v1):
        model_akida = convert(model_quantized)
        
        print("\nAkida Model Summary:")
        model_akida.summary()
    
    # Step 5: Map to Akida hardware
    print("\nStep 5: Detecting and mapping to Akida hardware...")
    devices = akida.devices()
    
    if len(devices) == 0:
        print("❌ No Akida device found!")
        print("   Model created but cannot be deployed to hardware.")
        print("   Running in software simulation mode only.")
        return
    
    device = devices[0]
    print(f"✅ Found Akida device: {device.desc}")
    print(f"   Device version: {device.version}")
    print(f"   Model IP version: {model_akida.ip_version}")
    
    # Map to hardware
    # model_akida.map(device)
    model_akida.map(device, hw_only=True)
    print("\n✅ Model mapped to hardware!")
    
    print("\nHardware Allocation:")
    model_akida.summary()
    
    # Step 6: Benchmark on Akida hardware (uses uint8)
    print("\nStep 6: Benchmarking on Akida hardware...")
    akida_results = benchmark_akida_hardware(model_akida, device, samples_uint8)
    
    # Step 7: Save the Akida model
    print("\nStep 7: Saving Akida model...")
    output_file = "simplecnn2_akida_v1_small.fbz"
    model_akida.save(output_file)
    file_size = Path(output_file).stat().st_size / 1024
    print(f"✅ Model saved to: {output_file}")
    print(f"   File size: {file_size:.2f} KB")
    
    # Summary comparison with actual measured values
    print("\n" + "="*70)
    print("PERFORMANCE SUMMARY - HARDWARE ACCELERATED")
    print("="*70)
    print(f"\n📊 KERAS BASELINE (CPU/GPU):")
    print(f"   Inference time: {keras_time:.3f} ± {keras_std:.3f} ms")
    print(f"   Throughput:     {1000/keras_time:.1f} fps")
    
    print(f"\n🔌 AKIDA HARDWARE:")
    print(f"   Batch inference:  {akida_results['batch_time_ms']:.3f} ms per image")
    print(f"   Single inference: {akida_results['single_time_ms']:.3f} ± {akida_results['single_std_ms']:.3f} ms")
    print(f"   Throughput:       {akida_results['batch_fps']:.1f} fps (batch mode)")
    
    print(f"\n⚡ POWER & ENERGY:")
    print(f"   Floor power:      {akida_results['floor_power_mw']:.2f} mW")
    print(f"   Average power:    {akida_results['avg_power_mw']:.2f} mW")
    print(f"   Energy/inference: {akida_results['energy_mj']:.2f} mJ")
    print(f"   Energy efficiency: {1000/akida_results['energy_mj']:.0f} inferences/joule")
    
    print(f"\n✅ Conv layers should be HARDWARE accelerated (check summary above)")
    print(f"   Input: {input_shape} fits Akida v1 constraint (width ≤ 256)")
    print("="*70)
    
    print("\n✅ Deployment complete!")
    
    return {
        'keras_time_ms': keras_time,
        'keras_std_ms': keras_std,
        'akida_results': akida_results,
        'input_shape': input_shape
    }


def compare_both():
    """
    Run both versions and compare results
    Call this in your notebook to see the difference between software and hardware acceleration
    """
    print("\n" + "🔴"*35)
    print("RUNNING LARGE INPUT VERSION (SOFTWARE MODE)")
    print("🔴"*35 + "\n")
    
    results_large = main_large_input()
    
    print("\n\n" + "🟢"*35)
    print("RUNNING SMALL INPUT VERSION (HARDWARE MODE)")
    print("🟢"*35 + "\n")
    
    results_small = main_small_input()
    
    # Final comparison
    print("\n" + "="*70)
    print("FINAL COMPARISON: SOFTWARE vs HARDWARE")
    print("="*70)
    
    print("\n📊 LARGE INPUT (480x440) - SOFTWARE MODE:")
    print(f"   Keras:  {results_large['keras_time_ms']:.2f} ms")
    print(f"   Akida:  {results_large['akida_results']['batch_time_ms']:.2f} ms (batch)")
    print(f"   Power:  {results_large['akida_results']['avg_power_mw']:.2f} mW")
    print(f"   Energy: {results_large['akida_results']['energy_mj']:.2f} mJ")
    
    print("\n📊 SMALL INPUT (240x220) - HARDWARE MODE:")
    print(f"   Keras:  {results_small['keras_time_ms']:.2f} ms")
    print(f"   Akida:  {results_small['akida_results']['batch_time_ms']:.2f} ms (batch)")
    print(f"   Power:  {results_small['akida_results']['avg_power_mw']:.2f} mW")
    print(f"   Energy: {results_small['akida_results']['energy_mj']:.2f} mJ")
    
    speedup = results_large['akida_results']['batch_time_ms'] / results_small['akida_results']['batch_time_ms']
    energy_improvement = results_large['akida_results']['energy_mj'] / results_small['akida_results']['energy_mj']
    
    print("\n🚀 IMPROVEMENTS:")
    print(f"   Speedup:         {speedup:.2f}x faster with hardware acceleration")
    print(f"   Energy savings:  {energy_improvement:.2f}x more efficient")
    
    print("="*70)
    
    return {
        'large': results_large,
        'small': results_small,
        'speedup': speedup,
        'energy_improvement': energy_improvement
    }
```

```python
def main():
    """
    Convenience function - runs the optimized small input version
    For comparison, call:
      - main_large_input() for software mode
      - main_small_input() for hardware mode
      - compare_both() to run both and see comparison
    """
    return main_small_input()
```

```python
if __name__ == "__main__":
    main()
```
